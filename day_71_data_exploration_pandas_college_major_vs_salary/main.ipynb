```python
import requests
from bs4 import BeautifulSoup
```

```python
def chunks(lst, n):
    """Yield successive n-sized chunks from lst."""
    for i in range(0, len(lst), n):
        yield lst[i:i + n]
```

```python
def dataframe_scrape(i):
    url = f'https://www.payscale.com/college-salary-report/majors-that-pay-you-back/bachelors/page/{i}'
    response = requests.get(url)
    website_html = response.text
    
    soup = BeautifulSoup(website_html, "html.parser")
    
    info = soup.find_all(class_="data-table__value")

    info_2 = []
    for item in info:
        info_2.append(item.text)
        
    new_info = chunks(info_2, 6)
    
    data_list = []
    for item in new_info:
        data_list.append(item)
        
    new_df = pd.DataFrame(data_list, columns=['Rank', 
                                          'Major', 
                                          'Degree', 
                                          'Early_Career_Pay',
                                         'Mid_Career_Pay',
                                         'High_Meaning'])
    return(new_df)
  
```

```python
website = []
for i in range(1, 34):
    website.append(dataframe_scrape(i))
result = pd.concat(website)
result.reset_index()
```

```python
finance = result[result['Major'] == 'Finance']
finance['Mid_Career_Pay']
```
